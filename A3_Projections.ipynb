```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_grouped = df.groupby('release year').agg({
    'numCharactersSUM_Male': 'sum',
    'numCharactersSUM_Female': 'sum',
    'numCharactersSUM_TANG+': 'sum'
}).reset_index()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_grouped['release year'], df_grouped['male_proportion'], marker='o', label='Male')
plt.plot(df_grouped['release year'], df_grouped['female_proportion'], marker='o', label='Female')
plt.plot(df_grouped['release year'], df_grouped['neutral_proportion'], marker='o', label='Neutral')

plt.title('Estimated Gender Equality in Games Over Time')
plt.xlabel('Release Year')
plt.ylabel('Proportion of Characters')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('genderPCgames_pythontest_v3.csv')
```

```python
df['release_date_new'] = pd.to_datetime(df['release_date_new'], format='%d/%m/%Y')
```

```python
df['release year'] = df['release_date_new'].dt.year
```

```python
df_grouped = df.groupby('release year').agg({
    'numCharactersSUM_Male': 'sum',
    'numCharactersSUM_Female': 'sum',
    'numCharactersSUM_TANG+': 'sum'
}).reset_index()
```

```python
df_grouped['total_characters'] = df_grouped['numCharactersSUM_Male'] + df_grouped['numCharactersSUM_Female'] + df_grouped['numCharactersSUM_TANG+']
```

```python
df_grouped['male_proportion'] = df_grouped['numCharactersSUM_Male'] / df_grouped['total_characters']
df_grouped['female_proportion'] = df_grouped['numCharactersSUM_Female'] / df_grouped['total_characters']
df_grouped['neutral_proportion'] = df_grouped['numCharactersSUM_TANG+'] / df_grouped['total_characters']
```

```python
X = df_grouped['release year'].values.reshape(-1, 1)
y_male = df_grouped['male_proportion'].values
y_female = df_grouped['female_proportion'].values
y_neutral = df_grouped['neutral_proportion'].values

reg_male = LinearRegression().fit(X, y_male)
reg_female = LinearRegression().fit(X, y_female)
reg_neutral = LinearRegression().fit(X, y_neutral)
```

```python
future_years = np.arange(2024, 3001).reshape(-1, 1)
predicted_male = reg_male.predict(future_years)
predicted_female = reg_female.predict(future_years)
predicted_neutral = reg_neutral.predict(future_years)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_grouped['release year'], df_grouped['male_proportion'], marker='o', label='Male (1984-2023)')
plt.plot(df_grouped['release year'], df_grouped['female_proportion'], marker='o', label='Female (1984-2023)')
plt.plot(df_grouped['release year'], df_grouped['neutral_proportion'], marker='o', label='Neutral (1984-2023)')
```

```python
plt.plot(future_years, predicted_male, linestyle='--', color='blue', label='Male (2024-3000)')
plt.plot(future_years, predicted_female, linestyle='--', color='orange', label='Female (2024-3000)')
plt.plot(future_years, predicted_neutral, linestyle='--', color='green', label='Neutral (2024-3000)')

plt.title('Estimated Gender Equality in Games Over Time')
plt.xlabel('Release Year')
plt.ylabel('Proportion of Characters')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.xticks(range(1980, 3001, 50), rotation=45)  # Rotate x-axis labels for better readability
plt.xlim(1980, 3000)  # Set x-axis limits from 1980 to 3000
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('genderPCgames_pythontest_v3.csv')
```

```python
df_grouped = df.groupby('release year').agg({
    'wordsSUM_male': 'sum',
    'wordsSUM_female': 'sum'
}).reset_index()
```

```python
df_grouped['total_words'] = df_grouped['wordsSUM_male'] + df_grouped['wordsSUM_female']
```

```python
df_grouped['male_percentage'] = (df_grouped['wordsSUM_male'] / df_grouped['total_words']) * 100
df_grouped['female_percentage'] = (df_grouped['wordsSUM_female'] / df_grouped['total_words']) * 100
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_grouped['release year'], df_grouped['male_percentage'], marker='o', label='Male Character Word Count')
plt.plot(df_grouped['release year'], df_grouped['female_percentage'], marker='o', label='Female Character Word Count')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_grouped['release year'], df_grouped['male_percentage'], marker='o', label='Male Character Word Count')
plt.plot(df_grouped['release year'], df_grouped['female_percentage'], marker='o', label='Female Character Word Count')

plt.title('Trends in Character Word Count Percentages (Gender-wise) Over Time')
plt.xlabel('Release Year')
plt.ylabel('Percentage of Total Word Counts')
plt.legend()
plt.grid(True)

# Add titles to the axes
plt.gca().set_xlabel('Release Year')
plt.gca().set_ylabel('Percentage of Total Word Counts')

# Add legend title
plt.gca().legend().set_title('Gender')

plt.tight_layout()
plt.xticks(range(1980, 2024, 5), rotation=45)  # Rotate x-axis labels for better readability
plt.xlim(1980, 2023)  # Set x-axis limits from 1980 to 2023
plt.ylim(0, 100)  # Set y-axis limits from 0 to 100 (percentages)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('genderPCgames_pythontest_v3.csv')
```

```python
df_grouped = df.groupby('release year').agg({
    'wordsSUM_male': 'sum',
    'wordsSUM_female': 'sum'
}).reset_index()
```

```python
df_grouped['total_words'] = df_grouped['wordsSUM_male'] + df_grouped['wordsSUM_female']
```

```python
df_grouped['male_percentage'] = (df_grouped['wordsSUM_male'] / df_grouped['total_words']) * 100
df_grouped['female_percentage'] = (df_grouped['wordsSUM_female'] / df_grouped['total_words']) * 100
```

```python
X = df_grouped['release year'].values.reshape(-1, 1)
y_male = df_grouped['male_percentage'].values
y_female = df_grouped['female_percentage'].values

reg_male = LinearRegression().fit(X, y_male)
reg_female = LinearRegression().fit(X, y_female)

# Project the trend into the future (2024-3000)
future_years = np.arange(2024, 3001).reshape(-1, 1)
predicted_male = reg_male.predict(future_years)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_grouped['release year'], df_grouped['male_percentage'], marker='o', label='Male Character Dialogue')
plt.plot(df_grouped['release year'], df_grouped['female_percentage'], marker='o', label='Female Character Dialogue')

# Plot the projected trend
plt.plot(future_years, predicted_male, linestyle='--', color='grey', label='Male Character Dialogue (2024-3000)')
plt.plot(future_years, predicted_female, linestyle='--', color='orange', label='Female Character Dialogue (2024-3000)')

plt.title('Trends in Character Word Count Percentages (Gender-wise) Over Time')
plt.xlabel('Release Year')
plt.ylabel('Percentage of Total Word Count')
plt.legend()
plt.grid(True)

# Add titles to the axes
plt.gca().set_xlabel('Release Year')
plt.gca().set_ylabel('Percentage of Total Word Count')

# Add legend title
plt.gca().legend().set_title('Gender')

plt.tight_layout()
plt.xticks(range(1984, 2050, 100), rotation=45)  # Rotate x-axis labels for better readability
plt.xlim(1984, 2050)  # Set x-axis limits from 2024 to 3000
plt.ylim(0, 100)  # Set y-axis limits from 0 to 100 (percentages)
plt.show()
```
